```python
# มาตรฐานข้อมูล
data_standardized = (df - df.mean()) / df.std()

# ทำ PCA
pca = PCA()
principal_components = pca.fit_transform(data_standardized)

# คำนวณความสัมพันธ์ของแต่ละ Principal Component
explained_variance_ratio = pca.explained_variance_ratio_

# Scree Plot
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()

# Loading Plot
loading_vectors = pca.components_

# แสดง Loading Plot ของ Principal Component 1 และ Principal Component 2
plt.figure(figsize=(10, 6))
plt.scatter(loading_vectors[0, :], loading_vectors[1, :])
plt.xlabel('Loading Vector PC1')
plt.ylabel('Loading Vector PC2')

# นำเสนอตัวแปรต้นทุนที่สำคัญที่สุดบนกราฟ
for i, txt in enumerate(range(1, len(loading_vectors[0, :]) + 1)):
    plt.annotate(txt, (loading_vectors[0, i], loading_vectors[1, i]))

plt.title('Loading Plot')
plt.show()
```
